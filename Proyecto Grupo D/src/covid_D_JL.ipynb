```python
import pandas as pd 
import json 
```

```python
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
covid_general = pd.read_csv(url, sep=",")
covid_general
```

```python
paises_grupoD=["PRT", "VEN", "TUR", "GBR", "ESP"]
covid_grupoD=covid_general[covid_general.iso_code.isin(paises_grupoD)]
covid_grupoD
```

```python
covid_grupoD.info()# queriamos ver el tipo Dtype de cada columna
```

```python
covid_grupoD.info()
```

```python
covid_grupoD.set_index(["date"],inplace=True)
```

```python
covid_grupoD
```

```python
covid_no_NaN = covid_grupoD.copy()
```

```python
covid_no_NaN.shape
```

```python
covid_no_NaN.dropna(subset=["total_deaths"],axis=0, inplace=True)
```

```python
covid_no_NaN
```

```python
final_covid_D = covid_no_NaN.groupby("date").mean().total_deaths
final_covid_D
```

```python
final_covid_D=final_covid_D.to_frame()
```

```python
final_covid_D.rename(columns={"total_deaths":"t_d_averages"},inplace=True)
final_covid_D
```
